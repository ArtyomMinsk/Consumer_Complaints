```python
consumer_complaints = pd.read_csv('complaints_dec_2014.csv')
```

```python
consumer_complaints
```

```python
consumer_complaints.describe()
```

```python
complaints = consumer_complaints['Product'].value_counts()
complaints.head()
```

```python
complaints.plot(title='Number of complaints by product', kind='bar', rot=90)
```

```python
complaints_by_company = consumer_complaints['Company'].value_counts().head(10)
complaints_by_company
```

```python
complaints_by_company.plot(title='Number of complaints by Company', kind='bar', rot=90)
```

```python
complaints_by_response = consumer_complaints['Company response'].value_counts()
complaints_by_response
```

```python
complaints_by_response.plot(title='Number of complaints by Company response', kind='bar', rot=90)
```

```python
consumer_complaints['Weekday'] = pd.to_datetime(consumer_complaints['Date received']).map(lambda d: d.weekday())
consumer_complaints.head()
```

```python
my_mean = consumer_complaints.groupby(['Weekday', 'Date received']).size().groupby(level=0).mean()
my_mean
```

```python
my_mean.plot(title='Mean number of complaints by day of week', kind='bar', rot=90)
```
