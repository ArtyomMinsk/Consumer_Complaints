```python
consumer_complaints = pd.read_csv('complaints_dec_2014.csv')
```

```python
consumer_complaints
```

```python
consumer_complaints.describe()
```

```python
complaints = consumer_complaints['Product'].value_counts()
complaints.head()
```

```python
complaints.plot(title='Number of complaints by product', rot=90)
```

```python
complaints_by_company = consumer_complaints['Company'].value_counts().head(10)
complaints_by_company
```

```python
complaints_by_company.plot(title='Number of complaints by Company', rot=90)
```

```python
complaints_by_response = consumer_complaints['Company response'].value_counts()
complaints_by_response
```

```python
complaints_by_response.plot(title='Number of complaints by Company response', rot=90)
```

```python
date = consumer_complaints.pop('Date sent to company')
```

```python
consumer_complaints.index = pd.to_datetime(date)
```

```python
consumer_complaints['Weekday'] = consumer_complaints.index.to_series().map(lambda d: d.weekday())
```
